```python
# Section 11.1
def hours(): # This function would be inside of the 'zoo.py' file. 
    print("Open 9-5 Daily") 

# Section 11.2
import zoo as menagerie # Imports the 'zoo.py' file if it were to exist. Aliased as 'menagerie' as instructed.

menagerie.hours() # Uses alias to call the hours function which prints the Zoo's hours.

# Section 16.8

# Imports used for 16.8 
import sqlalchemy as sa
from sqlalchemy.orm import declarative_base
from sqlalchemy.orm import sessionmaker

# SQLAlchemy 
# Most of this code is straight from the book and edited to work with the 'Books' example.
# It works perfectly for me in VSCode, but probably not in the Jupiter Notebook.

conn = sa.create_engine("sqlite:///books.db") # Connects to the sqlite books.db database
Base = declarative_base() # I think this is used to help create the database (not 100% sure)

class Books(Base):
    """
    This class is basically the database for Books in Python form.
    Creates the name for the table and the information Columns.
    """
    __tablename__ = "books"
    title = sa.Column("Title", sa.String, primary_key=True) # Primary key used to identify each book individually
    author = sa.Column("Author", sa.String)
    year = sa.Column("Year", sa.Integer)

    def __init__(self, title, author, year):
        """
        Function used to define things about the books in the database.
        """
        self.title = title
        self.author = author
        self.year = year
    def __repr__(self):
        """
        This function formats the information about each Book and returns it.
        """
        return "<Books({}, {}, {})>".format(self.title, self.author, self.year) 
    
Base.metadata.create_all(conn) # According to the textbook, this line "magically creates the database and table."

# Each book variable calls the Books class and makes its own book with its own data
book1 = Books("The Weirdstone of Brisingamen", "Alan Garner", 1960)
book2 = Books("Perdido Street Station", "China Mieville", 2000)
book3 = Books("Thud!", "Terry Pratchett", 2005)
book4 = Books("The Spellman Files", "Lisa Lutz", 2007)
book5 = Books("Small Gods", "Terry Pratchett", 1992)

# Creates a session to communicate with the database
Session = sessionmaker(bind=conn)
session = Session()
session.add_all([book1, book2, book3, book4, book5]) # Calls the book variables and adds the information to the database

# Queries for the books titles and then prints them
results = session.query(Books.title).all()
for title in results:
    print(title)
```
